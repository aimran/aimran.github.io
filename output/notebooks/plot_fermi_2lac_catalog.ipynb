## Create a fits file parser

We will create a small parser to quickly and efficiently read the fits file. It would be extremely tedious to have to open fits file, define tables, columns etc. Remember! If you are typing up the same thing over and over again, there is probably a better way to do things. 

With that in mind, lets look at parts of FLACatalog.py (available at my github repo). We'll only focus on the important bits --  Its essentially a class that encapsulates the opening of the fits file and defines a bunch of handy getters to extract the info from relevant columns.

```python
class FLACatalog:
    """
    Parse the FLAC fits file and extract data from the main table.
    """
    def __init__(self, filename):
        hdulist = pyfits.open(filename)
        tbl = hdulist[1].data  # main table
        try:
            self._sn      = tbl.field("Fermi_Name")
            self._assoc   = tbl.field("CName")
            self._ra      = tbl.field("RAJ2000") * Units.degree
            self._de      = tbl.field("DEJ2000") * Units.degree
            self._l2      = tbl.field("GLON") * Units.degree
            self._b2      = tbl.field("GLAT") * Units.degree
            self._an      = tbl.field("AngSep") * Units.degree
            self._an95    = tbl.field("PosErr95deg") * Units.degree
            self._class1  = tbl.field("Class")
            self._z       = tbl.field("Redshift")
            self._sed     = tbl.field("SED_Type")
            self._bayprob = tbl.field("BayProb")
            self._rgprob  = tbl.field("LR_RGProb")
            self._xgprob  = tbl.field("LR_XGProb")
            self._nsprob  = tbl.field("LogNLogSProb")
            self._rg      = tbl.field("LR_RG")
            self._xg      = tbl.field("LR_XG")
            self._F1e3    = tbl.field("Flux1000_100000") * Units.inv_cm2s
            self._dF1e3   = tbl.field("Unc_Flux1000_100000") * Units.inv_cm2s
            self._PIdx    = tbl.field("Spectral_Index")
            self._dPIdx   = tbl.field("Unc_Spectral_Index")
            self._ts      = tbl.field("TS")
            self._radFl   = tbl.field("Radio_Flux")
            self._xrayFl  = tbl.field("Xray_Flux")
            self._usnov   = tbl.field("USNOVmag")
            self._sdss    = tbl.field("SDSSVmag")
            self._aro     = tbl.field("ARO")
            self._aox     = tbl.field("AOX")
            self._clean   = tbl.field("CLEAN")
            self._in2fgl  = tbl.field("IN_2FGL")

            self._lgnupeak = self._logNuPeak(self.aox, self.aro)

        except KeyError as detail:
            print detail
            pass

        @property
        def ra(self):
            """
            List of J2000 right ascensions for the counterpart.
            """
            return self._ra
```

Lets deconstruct the class FLACatalog. It accepts a filename as an argument that should correspond to the fits file we created earlier. The <code>pyfits.open</code> function opens the file and returns a list of header/data units (hdulist). Our main table is stored as the second element <code>hdulist[1]</code> [first one corresponds to the main header of the file]. Afterwards, we associate different class variables to all the table columns in the fits file using the <code>field</code> method. And we are almost done!

All there is left to do is to set some convenient getters to grab the columns from the catalog on demand. For brevity, I'm only showing the <code>ra</code> method but hopefully you get the picture.

One word of caution. There are units built into several of the fields in the catalog. Therefore, I've tried my best to take care of them in the FLACatalog. You are encouraged to check/verify for yourself.

## Generate Plots

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

from FLACatalog import FLACatalog, Units
```

We instantiate an FLACatalog object by passing in the location of the fits file we created previously.

```python
catFLA = FLACatalog('../data/fermi_2lac_v01.fits')
```

#### Figure 2

We will begin by re-making figure 2 in the paper above. It shows the 95% containment radius as a function of TS (or the test statistic). The plot is further divided up into different population of AGN, namely, FSRQ, BL Lacs, unknown type, and other AGN. 

The first thing to do is to create a mask corresponding to the clean flag as set by the Fermi team. This is 

```python
#catalog name
catName = catFLA.catalog_name

#get the full array of source types or classes
#define masks based on the names present in the catalog 
#to cut on different types of sources
srcType = catFLA.source_class
fsrqFlg = srcType == 'FSRQ'
blacFlg = srcType == 'BL Lac'
agnFlg  = srcType == 'AGU'
aguFlg  = srcType == 'AGN'
radFlg  = srcType == 'Radio Galaxy'
ssrqFlg = srcType == 'SSRQ'
stbFlg  = srcType == 'Starburst galaxy'
unidFlg = srcType == 'Unidentified'
otherFlg = (aguFlg | radFlg | ssrqFlg | stbFlg | unidFlg)

#Test statistic
ts = catFLA.ts

#95% containment radius
errRad = catFLA.error_radius
```

Let the plotting fiesta begin! Please note that your plot may look slightly different from mine since I use a few custom style settings. 

```python
scatter_kwargs = dict(s=10, linewidths=0, edgecolors='none')

fig, ax = plt.subplots(figsize=(8,6))

ax.scatter(ts[fsrqFlg], errRad[fsrqFlg], color='r', label = 'FSRQ', **scatter_kwargs)
ax.scatter(ts[blacFlg], errRad[blacFlg], color='b', label = 'BL Lacs', **scatter_kwargs)
ax.scatter(ts[agnFlg], errRad[agnFlg], color='#00FF00', label = 'Unknown Type', **scatter_kwargs)
ax.scatter(ts[otherFlg], errRad[otherFlg], color='m', label = 'Other AGN', **scatter_kwargs)

ax.legend(scatterpoints = 1)

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_ylabel(u'95% Error radius [$^\circ$]')
ax.set_xlabel(u'TS')

plt.show()
```

#### Figure 6

Not bad! This looks very close to what you have on the paper. So, we are off to a good start. Lets plow on. We'll try figure 6 next. Its a plot of the distribution of log of synchrotron peak for 2 different species of AGN (FSRQ and BL Lac). Note that we are reusing mask from the previous along with a new clean mask as specified by the Fermi team.

```python
#clean flag
cleanFlg = catFLA.clean_flag

clean_fsrq = np.logical_and(cleanFlg==1, fsrqFlg)
lgnu_fsrq = catFLA.log_nu_peak[clean_fsrq]

clean_bllac = np.logical_and(cleanFlg==1, blacFlg)
lgnu_bllac = catFLA.log_nu_peak[clean_bllac]
```

```python
minorLocator = MultipleLocator(5)

fig, ax = plt.subplots(figsize=(8,6))
hist_kwargs = dict(bins=np.arange(12,17.5, 0.25), histtype='stepfilled', alpha=0.4)
ax.hist(lgnu_fsrq, color='r', label='FSRQ', **hist_kwargs)
ax.hist(lgnu_bllac, color='b', label='BL Lacs', **hist_kwargs)
ax.set_ylim(0,80)
ax.set_xlim(12,17.5)
ax.legend(numpoints=1)
ax.set_xlabel(r'$\log(\nu_{\rm peak}^{s} [{\rm Hz}])$')
ax.set_ylabel('Number of Sources')
ax.xaxis.set_minor_locator(minorLocator)
plt.show()
```

#### Figure 7

Once again its pretty close! In the next couple of panels, I will blast through some more plots. It should all be fairly straight-forward from here on. The basic idea is to use numpy's powerful array masking feature to select multiple species of AGN along with variable of interest. 

```python
cblac = np.logical_and(cleanFlg==1, blacFlg)
aoxb = catFLA.aox
arob = catFLA.aro
hspFlg = catFLA.sed_type == 'HSP'
ispFlg = catFLA.sed_type == 'ISP'
lspFlg = catFLA.sed_type == 'LSP'

scatter_kwargs = dict(s=10, linewidths=0, edgecolors='none')

fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(aoxb[cblac&hspFlg], arob[cblac&hspFlg], color='b', label = 'HSP', **scatter_kwargs)
ax.scatter(aoxb[cblac&ispFlg], arob[cblac&ispFlg], color='c', label = 'ISP', **scatter_kwargs)
ax.scatter(aoxb[cblac&lspFlg], arob[cblac&lspFlg], color='g', label = 'LSP', **scatter_kwargs)

ax.legend(scatterpoints = 1)

ax.set_ylabel(r'$\alpha_{\rm RO}$')
ax.set_xlabel(r'$\alpha_{\rm OX}$')
plt.show()
```

#### Figure 8

```python
lg = np.log10
cfsrq = np.logical_and(cleanFlg==1, fsrqFlg)

cblac = np.logical_and(cleanFlg==1, blacFlg)
radFlx = catFLA.radio_flux
xrayFlx = catFLA.xray_flux

hspFlg = catFLA.sed_type == 'HSP'
ispFlg = catFLA.sed_type == 'ISP'
lspFlg = catFLA.sed_type == 'LSP'

scatter_kwargs = dict(s=10, linewidths=0, edgecolors='none')

fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(lg(radFlx[cfsrq]), lg(xrayFlx[cfsrq]), color='r', label = 'FSRQ', **scatter_kwargs)
ax.scatter(lg(radFlx[cblac&hspFlg]), lg(xrayFlx[cblac&hspFlg]), color='b', label = 'HSP', **scatter_kwargs)
ax.scatter(lg(radFlx[cblac&ispFlg]), lg(xrayFlx[cblac&ispFlg]), color='c', label = 'ISP', **scatter_kwargs)
ax.scatter(lg(radFlx[cblac&lspFlg]), lg(xrayFlx[cblac&lspFlg]), color='g', label = 'LSP', **scatter_kwargs)

ax.legend(scatterpoints = 1)

ax.set_ylabel(r'$\log\ ({\rm X-ray\ flux\ [erg\ cm^{-2}\ sec^{-1}]})$')
ax.set_xlabel(r'$\log\ ({\rm S_{1.4} [mJy]})$')
plt.show()
```

#### Figure 10

This plot shows a normalized distribution of number of sources as a function of galactic latitude. Note I use the <code>weights</code> argument to specify the normalization factor.

```python
cfsrq = np.logical_and(cleanFlg==1, fsrqFlg)
cblac = np.logical_and(cleanFlg==1, blacFlg)
cunkwn = np.logical_and(cleanFlg==1,agnFlg)

#normalization factors
w1 = np.ones_like( catFLA.b2[cfsrq] )/catFLA.b2[cfsrq].size
w2 = np.ones_like(catFLA.b2[cblac]) / catFLA.b2[cblac].size
w3 = np.ones_like(catFLA.b2[cunkwn]) / catFLA.b2[cunkwn].size

hist_kwargs = dict(bins=np.arange(-1,1.1, 0.1), histtype='stepfilled', alpha=0.4)
fig, ax = plt.subplots(3, figsize=(8,8))


ax[0].hist(np.sin(catFLA.b2[cfsrq]), color='r', label='FSRQ', weights=w1,**hist_kwargs)
ax[0].legend(loc=9)
ax[1].hist(np.sin(catFLA.b2[cblac]), color='b', label='BL Lacs', weights=w2, **hist_kwargs)
ax[1].legend(loc=9)
ax[2].hist(np.sin(catFLA.b2[cunkwn]), color='g', label='Unknown', weights=w3, **hist_kwargs)
ax[2].legend(loc=9)

ax[1].set_ylabel(r'Normalized # of sources')
ax[2].set_xlabel(r'$\sin({\rm b})$')

for i in [0,1,2]:
    ax[i].set_ylim(0,0.13)

plt.tight_layout()
plt.show()
```

#### Figure 12

Okay, this is the last plot, I promise! I'll leave a few for you to figure out. This plot shows the number distribution of AGN as a function of redshift.

```python
cfsrq = np.logical_and(cleanFlg==1, fsrqFlg)
cblac = np.logical_and(cleanFlg==1, blacFlg)

hspFlg = catFLA.sed_type == 'HSP'
ispFlg = catFLA.sed_type == 'ISP'
lspFlg = catFLA.sed_type == 'LSP'
valid = catFLA.redshift > 0.

rs = catFLA.redshift

w1 = np.ones_like( rs[cfsrq] )/rs[cfsrq].size
w2 = np.ones_like(rs[cblac&valid]) / rs[cblac&valid].size
w3 = np.ones_like(rs[cblac&valid]) / rs[cblac].size
w3h = np.ones_like(rs[cblac&hspFlg&valid]) / rs[cblac&hspFlg&valid].size
w3i = np.ones_like(rs[cblac&ispFlg&valid]) / rs[cblac&ispFlg&valid].size
w3l = np.ones_like(rs[cblac&lspFlg&valid]) / rs[cblac&lspFlg&valid].size


hist_kwargs = dict(bins=np.arange(0,3.6, 0.1), histtype='step', alpha=1.)
fig, ax = plt.subplots(3, figsize=(8,8))

ax[0].hist(rs[cfsrq], color='r', label='FSRQ', weights=w1,**hist_kwargs)
ax[0].legend(loc=1)

ax[1].hist(rs[cblac&valid], color='b', weights=w2, label='BL Lacs',  **hist_kwargs)
ax[1].legend(loc=1)

ax[2].hist(rs[cblac&hspFlg&valid], color='b', weights=w3h, label='HSP', **hist_kwargs)
ax[2].hist(rs[cblac&ispFlg&valid], color='c', weights=w3i, label='ISP', **hist_kwargs)
ax[2].hist(rs[cblac&lspFlg&valid], color='g', weights=w3l, label='LSP', **hist_kwargs)
ax[2].legend(loc=1)

ax[1].set_ylabel(r'Normalized # of sources')
ax[2].set_xlabel(r'Redshift')

plt.tight_layout()
plt.show()
```
